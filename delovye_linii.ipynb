## Калькулятор на сайте (https://www.dellin.ru/requests/)

### Вариант с undetected_chromedriver

#### Получение необходимых параметров

```python
# настраиваем браузер
browser_options = uc.ChromeOptions()
browser_options.add_argument("--no-sandbox")
browser_options.add_argument("--disable-popup-blocking")
browser_options.add_argument("--disable-notifications")
browser_options.add_argument("--no-default-browser-check")
browser_options.add_argument("--disable-extensions")
browser_options.add_argument("--disable-dev-shm-usage")
browser_options.add_argument(f"user_agent={user_agent}")
browser_options.add_argument("--disable-blink-features=AutomationControlled")

# заходим на главную страницу
browser = uc.Chrome(options=browser_options, headless=False, use_subprocess=False, driver_executable_path=r".\chromedriver.exe")
browser.get("https://www.dellin.ru")
browser.minimize_window()
time.sleep(5)

# получаем сессию с cookies
session = requests.Session()
for cookie in browser.get_cookies():
    session.cookies.set(cookie["name"], cookie["value"], domain=cookie["domain"])
browser.quit()
```

#### Идентификация населённых пунктов

```python
url = "https://www.dellin.ru/api/cities/search.json"
headers = {"Accept": "application/json, text/javascript, */*; q=0.01",
           "Accept-Encoding": "gzip, deflate, br, zstd",
           "Accept-Language": "ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7",
           "User-Agent": user_agent}
city_name = "Казань"
region_name = "Татарстан"
input_data = {"q": " ".join((city_name, region_name)).strip(" ")}
resp = session.get(url=url, params=input_data, headers=headers)
resp.raise_for_status()
output_data = resp.json()
output_data
```

#### Идентификация населённых пунктов

```python
url = "https://188.68.188.69/api/cities/search.json"
headers = {"Accept": "application/json, text/javascript, */*; q=0.01",
           "Host": "www.dellin.ru",
           "Upgrade-Insecure-Requests": "1",
           "Accept-Encoding": "gzip, deflate, br, zstd",
           "Accept-Language": "ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7",
           "User-Agent": user_agent}
city_name = "Казань"
region_name = "Татарстан"
input_data = {"q": " ".join((city_name, region_name)).strip(" ")}
resp = requests.get(url=url, params=input_data, headers=headers, verify=False)
resp.raise_for_status()
output_data = resp.json()
output_data
```
