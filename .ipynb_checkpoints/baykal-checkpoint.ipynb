## Мини-калькулятор (https://www.baikalsr.ru/)

```python
# идентификация населённых пунктов
url = "https://www.baikalsr.ru/json/api_fias_cities_v2.json"
headers = {"Accept": "application/json, text/javascript, */*; q=0.01",
           "Accept-Encoding": "gzip, deflate, br, zstd",
           "Accept-Language": "ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7",
           "X-Requested-With": "XMLHttpRequest",
           "User-Agent": user_agent}
city_name = "Орёл"
region_name = "Орловская"
city_name = city_name.lower().strip(" ")
region_name = region_name.lower().strip(" ")
input_data = {"text": " ".join((city_name, region_name))}
resp = requests.get(url=url, params=input_data, headers=headers)
resp.raise_for_status()
output_data = resp.json()
output_data
```

```python
# расчёт стоимости и сроков доставки
url = "https://www.baikalsr.ru/json/alter_calc_v2.php"
headers = {"Accept": "*/*",
           "Accept-Encoding": "gzip, deflate, br, zstd",
           "Accept-Language": "ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7",
           "X-Requested-With": "XMLHttpRequest",
           "X-Kl-Saas-Ajax-Request": "Ajax_Request",
           "User-Agent": user_agent}
input_data = {"transport[]": "auto",
              "transport[]": "mgg",
              "transit_days": 1,
              "departure[cityguid]": "93b3df57-4c89-44df-ac42-96f05e9cd3b9",
              "destination[cityguid]": "dd8caeab-c685-4f2a-bf5f-550aca1bbc48",
              "cargo[summarycargo][weight]": 5,
              "cargo[summarycargo][volume]": 0.01,
              "cargo[summarycargo][units]": 1}
resp = requests.get(url=url, params=input_data, headers=headers)
resp.raise_for_status()
output_data = resp.json()
output_data
```
