## Калькулятор на сайте франчайзи (https://cdek-calc.ru/)

```python
# идентификация населённых пунктов
url = "https://api.cdek.ru/city/getListByTerm/jsonp.php"
headers = {"Accept": "*/*",
           "Accept-Encoding": "gzip, deflate, br",
           "Accept-Language": "ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7",
           "User-Agent": user_agent}
city_name = "Набережные Челны"
input_data = {"callback": "jQuery33108627187275869952_1723114023637",
              "q": city_name.split(" ", 1)[0],  # строка не должна содержать пробелов
              "name_startsWith": city_name.split(" ", 1)[0],
              "_": "1723114023644"}
resp = requests.get(url=url, params=input_data, headers=headers, verify=False)
output_data = json.loads(resp.text.replace("jQuery33108627187275869952_1723114023637", "")[5:-2])
output_data["geonames"]
```

```python
# расчёт стоимости и сроков доставки
headers = {"Accept": "*/*",
           "Accept-Encoding": "gzip, deflate, br, zstd",
           "Accept-Language": "ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7",
           "Connection": "keep-alive",
           "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
           "X-Requested-With": "XMLHttpRequest",
           "User-Agent": user_agent}
url = "https://cdek-calc.ru/calc/result"
input_data = {"city_0": 433,
              # "city_name_0": "Набережные Челны",
              "city_1": 44,
              # "city_name_1": "Москва",
              "goods[0][weight]": 5.5,
              "goods[0][length]": 1,
              "goods[0][width]": 1,
              "goods[0][height]": 1,
              "cena": 0,  # объявленная ценность в рублях
              "str": 0,  # страховка (0 - нет, 1 - да)
              "nal": 0,  # наложенные платёж (0 - нет, 1 - да)
              "otpr": 1,  # доставку оплачивает отправитель (0 - нет, 1 - да)
              "dogovor": 2}  # тип договора (1 - договор интернет-магазина, 2 - нет договора)
resp = requests.post(url=url, data=input_data, headers=headers)
output_data = {}
soup = BeautifulSoup(resp.text, "lxml")
rate_tags = soup.find_all("tr", class_="tarif_tr")
for rate_tag in rate_tags:
    rate_name = rate_tag.find("td", attrs={"align": "left"}).text.strip("\n").split("\n",1)[0]
    price = re.search(r"\d+(\.|,)*\d+(\.\d+)*", rate_tag.find_all("td", attrs={"align": "center"})[0].text).group(0).replace(",", "")
    terms = re.findall(r"\d+", rate_tag.find_all("td", attrs={"align": "center"})[3].text)
    if len(terms) == 2:
        term_min = terms[0]
        term_max = terms[1]
    else:
        term_min = terms[0]
        term_max = term_min
    output_data.update({rate_name: {"price": price,
                                    "term_min": term_min,
                                    "term_max": term_max}})
output_data
```
