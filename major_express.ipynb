```python
import requests
import json
from suds.client import Client
from suds.cache import NoCache
from decimal import Decimal
from bs4 import BeautifulSoup
import re
```

```python
# вспомогательная функция для удобного поиска необходимых входных параметров
def ParamsStrToDict(params_str: str) -> dict:
    res = {}
    for el in params_str.split("\n")[1:-1]:
        res.update({el.split(":", 1)[0]: el.split(":", 1)[1].strip()})
    return res
```

## Публичный API (https://developers.major-express.ru/)

```python
def SudsToDict(obj) -> dict:
    """
    Конвертирование объекта suds в словарь.

    Параметры
    ----------
    obj :
        Объект suds.

    Результат
    -------
    dict
        Словарь с содержимым объекта suds.

    """

    if hasattr(obj, '__keylist__'):
        res = {}
        fields = obj.__keylist__
        for field in fields:
            val = getattr(obj, field)
            if isinstance(val, list):
                res[field] = []
                for item in val:
                    res[field].append(SudsToDict(item))
            elif isinstance(val, Decimal):
                res[field] = float(val)
            else:
                res[field] = SudsToDict(val)
    else:
        res = obj
    return res
```

### Идентификация населённых пунктов для экспресс-доставки

```python
# см. https://developers.major-express.ru/web-servisy-2/web-servis-klientskij/ekspress-dostavka/metody-bez-avtorizaczii/
express_client = Client(url="https://ltl-ws.major-express.ru/ed.asmx?WSDL", cache=NoCache(), timeout=30, proxy=None)

# список населённых пунктов
# см. https://developers.major-express.ru/web-servisy-2/web-servis-klientskij/ekspress-dostavka/metody-bez-avtorizaczii/spisok-obsluzhivaemyh-gorodov/
express_res = {}
resp = express_client.service.dict_Cities()
for city in SudsToDict(resp)["EDCity"]:
    express_res.update({city["Code"]: city["Name"]})
express_res
```

### Расчёт стоимости и сроков доставки для экспресс-доставки

```python
# см. https://developers.major-express.ru/web-servisy-2/web-servis-klientskij/ekspress-dostavka/metody-bez-avtorizaczii/kalkulyator-stoimostej-i-srokov/
resp = express_client.service.Calculator(86, 111, 5000, 1)
express_res = SudsToDict(resp)
express_res
```
